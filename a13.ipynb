```python
import numpy as np
from collections import Counter
import pandas as pd
```

```python
with open("13/input.txt") as f:
    p = True
    pairs = []
    folds = []
    for line in f:
        if line == '\n': p = False
        elif p == True:
            (x, y) = line.split(',')
            pairs += [[int(x), int(y)]]
        else:
            (c, n) = line.strip()[11:].split('=')
            folds += [[c, int(n)]]

pairs = np.array(pairs)
(m, n) = np.amax(pairs, axis = 0)
print(m, n)
print(folds)
```

```python
array = np.zeros((m+1, n+1))
for p in pairs:
    array[p[0]][p[1]] = 1
```

```python
def foldx(a):
    b = np.zeros(a.shape)
    for i in range(a.shape[0]//2):
        b[i] = np.maximum(a[i], a[a.shape[0]-i-1])
    return b[:a.shape[0]//2,:]
                 
def foldy(a):
    b = np.zeros(a.shape)
    for i in range(a.shape[1]//2):
        b[:,i] = np.maximum(a[:,i], a[:,a.shape[1]-i-1])

    return b[:,:a.shape[1]//2]
                
for f in folds[:1]:
    array = foldx(array) if f[0] == 'x' else foldy(array)

print(Counter(array.ravel())[1])
for l in array:
    print(l, '\n')
```
