```python
import numpy as np
import pandas as pd
```

```python
with open("04/input.txt") as f:
    first = True
    for line in f:
        if first: 
            numbers = [int(x) for x in line.split(',')]
            first = False
            bingos = []
            continue
        if line == '\n':
            l = 1
            rows = []
            continue
        if l < 5:
            new_row = [int(x) for x in line.split()]
            rows.append(new_row)
            l += 1
        else:
            new_row = [int(x) for x in line.split()]
            rows.append(new_row)
            bingos.append(pd.DataFrame(rows))
print(len(numbers), numbers)
for df in bingos:
    print(df.head())
```

```python
def compute_last():
    last = 0
    s = 0
    for n in numbers:
        idx = 0
        while idx < len(bingos):
            df = bingos[idx]
            df[df == n] = np.NaN
            if any(df.isnull().apply(lambda x: all(x), axis=0)) or\
                any(df.isnull().apply(lambda x: all(x), axis=1)):
                    s = df.sum().sum(axis=0)
                    last = n * s
                    bingos.pop(idx)
                    idx -= 1
            idx += 1
    return last

compute_last()
```
